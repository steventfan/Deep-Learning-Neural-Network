```python
import gzip
import matplotlib.pyplot as plt
import numpy
import struct
import urllib.request
from matplotlib.lines import Line2D
from scipy import stats

def getData(fileName):
    file = gzip.open(fileName, 'r')
    file.read(4)
    numImages = struct.unpack('>i', file.read(4))[0]
    numRows = struct.unpack('>i', file.read(4))[0]
    numColumns = struct.unpack('>i', file.read(4))[0]
    
    buffer = file.read(numImages * numRows * numColumns)
    data = numpy.frombuffer(buffer, dtype = numpy.uint8)
    data = data.reshape(numImages, numRows * numColumns)

    return data, numRows * numColumns

def getLabels(fileName):
    file = gzip.open(fileName, 'r')
    file.read(4)
    numImages = struct.unpack('>i', file.read(4))[0]
    
    buffer = file.read(numImages)
    data = numpy.frombuffer(buffer, dtype = numpy.uint8)
    
    return data

def linearClassifier(trainData, trainLabels, testData, testLabels, weightRowSize, learningRate, batchSize):
    if len(trainData) == 0:
        return 0
    
    W = numpy.zeros((weightRowSize, 1))
    accuracies = []
    for i in range(1000):
        indices = numpy.random.choice(len(trainData), batchSize, replace = False)
        X = trainData[indices]
        Y = trainLabels[indices].reshape(-1, 1)
        
        W -= learningRate * (1 / batchSize) * (X.T.dot(X.dot(W)) - X.T.dot(Y))
            
        accuracy = 0
        if len(testData) > 0:
            for testImage, testLabel in zip(testData, testLabels):
                predictedLabel = testImage.dot(W)
                if ((predictedLabel <= 0.5 and testLabel == 0) or (predictedLabel > 0.5 and testLabel == 1)):
                    accuracy += 1
            accuracy /= len(testData)
        accuracies.append(accuracy)
        
    return accuracies

def neuralNetworkClassifier(function, trainData, trainLabels, testData, testLabels, weightRowSize, k, epochs, learningRate):
    V = numpy.random.normal(0, 0.01, (k, 1))
    W = numpy.random.normal(0, 1 / k, (k, weightRowSize))
    
    testAccuracies = []
    trainAccuracies = []
    for i in range(epochs):
        indices = numpy.random.choice(len(trainData), len(trainData), replace = False)
        X = trainData[indices]
        Y = trainLabels[indices]
        
        for j, (x, y) in enumerate(zip(X, Y)):
            val = W.dot(x)
            h = function(val, False).reshape(k, 1)
            derivativeH = function(val, True).reshape(k, 1)
            residual = V.T.dot(h).item() - y
            
            V -= learningRate * residual * h
            W -= learningRate * residual * numpy.multiply(V, derivativeH) * x
            
            if ((j % 1000 == 0) or ((i == epochs - 1) and (j == (len(trainData) - 1)))):
                accuracy = 0
                if len(testData) > 0:
                    for testImage, testLabel in zip(testData, testLabels):
                        predictedLabel = V.T.dot(function(W.dot(testImage), False))   
                        if ((predictedLabel <= 0.5 and testLabel == 0) or (predictedLabel > 0.5 and testLabel == 1)):
                            accuracy += 1
                            
                    accuracy /= len(testData)
                testAccuracies.append(accuracy)
                
                accuracy = 0
                for trainImage, trainLabel in zip(trainData, trainLabels):
                    predictedLabel = V.T.dot(function(W.dot(trainImage), False))
                    if ((predictedLabel <= 0.5 and trainLabel == 0) or (predictedLabel > 0.5 and testLabel == 1)):
                        accuracy += 1
                        
                accuracy /= len(trainData)
                trainAccuracies.append(accuracy)
        
    return testAccuracies, trainAccuracies
            
def sigmoid(val, isDerivative):
    x = ReLU(val, isDerivative)
    return 1 / (1 + numpy.exp(-x))
    
def ReLU(val, isDerivative):
    if isDerivative:
        return numpy.array([0 if i <= 0 else 1 for i in val])
    
    return numpy.maximum(val, 0)

learningRates = [10**-3 ,10**-4]
epochs = 5
K = [5, 40, 200]
N = 50000
colors = ['blue', 'green', 'red']

iterationVariables = numpy.append([1], numpy.arange(1000, (epochs * N) + 1, 1000))

legendHandles = []
for i, k in enumerate(K):
    legendHandles.append(Line2D([0], [0], marker = 'o', color = 'white', markerfacecolor = colors[i], label = 'k = ' + str(K[i])))
legendHandles.append(Line2D([0], [0], color = 'black', label = 'Test Dataset'))
legendHandles.append(Line2D([0], [0], linestyle = 'dashed', color = 'black', label = 'Train Dataset'))

print('Initialization complete')
```

```python
website = 'http://yann.lecun.com/exdb/mnist/'
files = ['train-images-idx3-ubyte.gz', 'train-labels-idx1-ubyte.gz', 't10k-images-idx3-ubyte.gz', 't10k-labels-idx1-ubyte.gz']
for file in files:
    url = website + file
    urllib.request.urlretrieve(url, file)

trainDataRaw, numElements = getData(files[0])
trainLabelRaw = getLabels(files[1])
testDataRaw, _ = getData(files[2])
testLabelRaw = getLabels(files[3])

print('File downloading and formatting complete')
```

```python
trainData = []
trainLabels = []
for i in range(len(trainDataRaw)):
    trainData.append(stats.zscore(trainDataRaw[i]))
    trainData[i] = numpy.append(trainData[i], 1)
    if trainLabelRaw[i] <= 4:
        trainLabels.append(0)
    else:
        trainLabels.append(1)

trainData = numpy.array(trainData)
trainLabels = numpy.array(trainLabels)
        
testData = []
testLabels = []
for i in range(len(testDataRaw)):
    testData.append(stats.zscore(testDataRaw[i]))
    testData[i] = numpy.append(testData[i], 1)
    if testLabelRaw[i] <= 4:
        testLabels.append(0)
    else:
        testLabels.append(1)

testData = numpy.array(testData)
testLabels = numpy.array(testLabels)

print('Normalization complete')
```

```python
if N < len(trainData):
    indices = numpy.random.choice(len(trainData), N, replace = False)
    trainDataSubset = trainData[indices]
    trainLabelSubset = trainLabels[indices]
else:
    trainDataSubset = trainData
    trainLabelSubset = trainLabels
    
accuracies = linearClassifier(trainDataSubset, trainLabelSubset, testData, testLabels, numElements + 1, learningRates[0], 100)

plt.plot(numpy.arange(1, 1001), accuracies)
plt.title('Accuracy vs Iteration')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.show()

print('Test accuracy for linear classifier:', accuracy)
```

```python
testAccuracies1 = []
trainAccuracies1 = []
for k in K:    
    testAccuracy, trainAccuracy = neuralNetworkClassifier(ReLU, trainDataSubset, trainLabelSubset, testData, testLabels, numElements + 1, k, epochs, learningRates[1])
    testAccuracies1.append(testAccuracy)
    trainAccuracies1.append(trainAccuracy)

for i in range(len(K)):
    plt.plot(iterationVariables, testAccuracies1[i], color = colors[i])
    plt.plot(iterationVariables, trainAccuracies1[i], color = colors[i], linestyle = 'dashed')
plt.title('Accuracy vs Iteration')
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.legend(handles = legendHandles)
plt.show()

for i, k in enumerate(K):
    print('Final average accuracy for k = ', k, ': ', testAccuracies1[i][-1], sep = '')
```
